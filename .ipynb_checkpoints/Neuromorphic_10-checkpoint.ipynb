###  Victor Purpura in Python

```python
def victor_purpura_dist(event_1,event_2,cost=1):
    d=0
#  d=spkd(tli,tlj,cost) calculates the "spike time" distance
#  (Victor & Purpura 1996) for a single cost

#  tli: vector of spike times for first spike train
#  tlj: vector of spike times for second spike train
#  cost: cost per unit time to move a spike

#   Copyright (c) 1999 by Daniel Reich and Jonathan Victor.
#   Translated to Matlab by Daniel Reich from FORTRAN code by Jonathan Victor.

    for channel in np.arange(1,16):
        tli=event_1[event_1[:,1]==channel,0]
        tlj=event_2[event_2[:,1]==channel,0]

        nspi=len(tli);
        nspj=len(tlj);

        scr=np.zeros([nspi+1,nspj+1])

        scr[:,0]=np.arange(0,nspi+1)
        scr[0,:]=np.arange(0,nspj+1)
        if nspi and nspj:
            for i in np.arange(1,nspi+1):
                for j in np.arange(1,nspj+1):
                    scr[i,j]=min(scr[i-1,j]+1, scr[i,j-1]+1, scr[i-1,j-1]+cost*abs(tli[i-1]-tlj[j-1]));
        d=d+scr[nspi,nspj]
    return d
```

```python
#Victor_purpura for every events_list
fig=plt.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')
dist_a=[]
dist_b=[]
dist_c=[]
dist_d=[]
dist=[dist_a,dist_b,dist_c,dist_d]
for probes in probe_events:
    for event_b in probes:
        dist_a.append(victor_purpura_dist(probe_a[0],event_b,0.0001))
        dist_b.append(victor_purpura_dist(probe_b[0],event_b,0.0001))
        dist_c.append(victor_purpura_dist(probe_c[0],event_b,0.0001))
        dist_d.append(victor_purpura_dist(probe_d[4],event_b,0.0001))
ax1=plt.subplot(221)
ax1.plot(dist_a,c='r')
ax1.hlines(max(dist_a[0:67]), 0, 210, colors='k', linestyles='solid', label='')
ax2=plt.subplot(222)
ax2.plot(dist_b,c='b')
ax2.hlines(max(dist_b[67:115]), 0, 210, colors='k', linestyles='solid', label='')
ax3=plt.subplot(223)
ax3.plot(dist_c,c='g')
ax3.hlines(max(dist_c[116:185]), 0, 210, colors='k', linestyles='solid', label='')
ax4=plt.subplot(224)
ax4.plot(dist_d,c='c')
ax4.hlines(max(dist_d[188:200]), 0, 210, colors='k', linestyles='solid', label='')

                          
                          
#                           67
# 49
# 70
# 24
```
